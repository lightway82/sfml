```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000

#TODO
def limitTransform(element):
    if element <= 10000:
        return "A"
    elif element <= 100000:
        return "B"
    elif element <= 200000:
        return "C"
    elif element <= 400000:
        return "D"
    elif element <= 700000:
        return "E"
    else:
        return "F" 
    
    
df['c_lim'] = df['LIMIT_BAL'].map(limitTransform)
df[['LIMIT_BAL', 'c_lim']].head(5)
```

```python
# (7) постироить распределение LIMIT_BAL (гистрограмму)
import matplotlib.pyplot as plt
import numpy as np

median = df["LIMIT_BAL"].median()
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1, 1, 1)
ax.set_title("Кредитный лимит гистограмма")
ax.set_xlabel("Частота")
ax.set_ylabel("Кредитный лимит")
ax.grid(True)
ax.annotate('Медиана', xy=(median, 6000), arrowprops=dict(arrowstyle='->'), xytext=(median, 8500))
ax.hist(df["LIMIT_BAL"], bins=df["LIMIT_BAL"].unique().sort(), facecolor='g', alpha=0.75)
```

```python
# (8) построить зависимость кредитного лимита от образования для каждого пола
# график необходимо сделать очень широким (на весь экран)
import numpy as np

educations = np.sort(df["EDUCATION"].unique())
sex1_df = df[["LIMIT_BAL", "SEX", "EDUCATION"]]
result = sex1_df.set_index(["EDUCATION"]).groupby(["SEX", "EDUCATION"]).median()

fig = plt.figure(figsize=(15, 9))
ax = fig.add_subplot(1, 1, 1)
ax.set_title("Кредитный лимит по образованию для поля 1")
ax.set_ylabel("Кредитный лимит")
ax.set_xlabel("Образование")


bar_width = 0.35
ax.bar(educations - bar_width/2, result.loc[1].values.flatten(), 0.32, alpha=0.85,   color='b',  label='SEX1')
ax.bar(educations + bar_width/2, result.loc[2].values.flatten(), 0.32, alpha=0.85,   color='g',  label='SEX2')

ax.set_xticks(educations)
ax.legend()
```
