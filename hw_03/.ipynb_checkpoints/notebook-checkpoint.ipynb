```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_curve, precision_recall_curve
from sklearn import preprocessing
from sklearn.utils import shuffle

numeric_attrs = ['age', 'duration', 'campaign', 'pdays', 'previous',
                     'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']
bin_attrs = ['default', 'housing', 'loan']
cate_attrs = ['poutcome', 'education', 'job', 'marital', 'contact', 'month','day_of_week']

def feature_scaling(data, numeric_attrs):
    """
    Стандартное масштабирование числовых фич, по mean и std
    """
    for i in numeric_attrs:
        std = data[i].std()
        if std != 0:
            data[i] = (data[i]-data[i].mean()) / std
        else:
            data = data.drop(i, axis=1)
    return data


def encode_edu_attrs(data):
    """
    Заменяет строку с образованием на его урове от 1 и до макс уровня
    """
    values = ["illiterate", "basic.4y", "basic.6y", "basic.9y", 
    "high.school",  "professional.course", "university.degree"]
    levels = range(1,len(values)+1)
    dict_levels = dict(zip(values, levels))
    for v in values:
        data.loc[data['education'] == v, 'education'] = dict_levels[v] 
        
    data['education'] = data.education.map(lambda v: v if v!='unknown' else 1)
    return  data       
            

def encode_cate_attrs(data, cate_attrs, encodeEducation=False):
    """Преобразует категоральные атрибуты, по желанию можно включть кодирование образования по уровням(станет числовым)"""
    cate_attrs_copy = cate_attrs.copy()
    if encodeEducation:
        data = encode_edu_attrs(data)
        cate_attrs_copy.remove('education')        
    
    return pd.get_dummies(data, columns=cate_attrs_copy)
    
def fill_unknown(data, bin_attrs, models=None):  
    """Заполнит unknown поля бинарных фич на основании предскзания
    по остальным данным, используя RandomForestClassifier
    [models] словарь модели по имени бинарного аттребута, это словать который был возвращен при выполнении на обучающей выборке 
    """
    fitted_models=dict()
    for i in bin_attrs:     
        test_data = data[data[i] == 'unknown']
        if 'target' in test_data.columns:
            testX = test_data.drop(bin_attrs+['target'], axis=1)
        else:
            testX = test_data.drop(bin_attrs, axis=1)
        train_data = data[data[i] != 'unknown']        
        trainY = train_data[i].astype('int32')
        if 'target' in train_data.columns:
            trainX = train_data.drop(bin_attrs+['target'], axis=1)
        else:
            trainX = train_data.drop(bin_attrs, axis=1)
        test_data[i], model = train_predict_unknown(trainX, trainY, testX,
                                                    model= models[i] if models is not None else None)
        fitted_models[i]=model
        data = pd.concat([train_data, test_data])    
    return data, fitted_models

def train_predict_unknown(trainX, trainY, testX, model=None):    
    if model is None:        
        forest = RandomForestClassifier(n_estimators=100)
        forest = forest.fit(trainX, trainY)
    else:
        forest = model    
    test_predictY = forest.predict(testX).astype(int)
    return pd.DataFrame(test_predictY, index=testX.index), forest
    

def encode_bin_attrs(data, bin_attrs, models=None):
    """Заменяет бинарные атрибуты yes no на 1 и 0. 
    Значения unknown заменит предсказанными
    Вернет кортеж из data и models - словарь моделей по атрибутам
    """
    for i in bin_attrs:
        data.loc[data[i] == 'no', i] = 0
        data.loc[data[i] == 'yes', i] = 1
    
    data, models = fill_unknown(data, bin_attrs, models)
    
    for i in bin_attrs:
        data[i] = data[i].astype("int32")
    
    return data, models

def trans_num_attrs(data, numeric_attrs, factorizeAge=False, facBins=10, scaling = False):
    """Масштабирует числовые атрибуты и по выбору факторизует возраст на группы"""
    if factorizeAge:        
        data["age"] = pd.qcut(data['age'], facBins)
        data["age"] = pd.factorize(data["age"])[0]+1 
        
    if scaling:
        return feature_scaling(data, numeric_attrs)
    else:
        return data
        
    


def preproc_data(df_input, *, factorizeAge, facAgeBins=10, encodeEducation, models_for_unknown=None, scaling=True): 
    df_output = df_input.copy()
    
    #if models_for_unknown is None:
    #    df_output = shuffle(df_input)       
         
        
    df_output.drop(["_id"], axis=1, inplace=True)   
      
    df_output = encode_cate_attrs(df_output, cate_attrs, encodeEducation)
    df_output, models = encode_bin_attrs(df_output, bin_attrs, models_for_unknown)
    
    numeric_attrs_ext = numeric_attrs.copy()
    if encodeEducation:
        numeric_attrs_ext += ['education']        
    df_output = trans_num_attrs(df_output, numeric_attrs_ext, factorizeAge=factorizeAge, facBins=facAgeBins, scaling=scaling)  
    
    return df_output, models


%matplotlib inline

def plot_roc_curve(y_test, predict):
    fpr, tpr, thresholds = roc_curve(y_test, predict)
    rc_score = roc_auc_score(y_test, predict)
    plt.title('ROC (AUC=%0.2f)' % (rc_score))  
    plt.fill_between(fpr, tpr, alpha=0.2)  
    plt.grid(True, linestyle='-', color='0.75')  
    plt.plot(fpr, tpr, linewidth=2, label="roc")
    plt.plot(fpr, thresholds, linewidth=1, label="thresholds")
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('fpr')
    plt.ylabel('tpr')
    plt.axis([0, 1, 0, 1])
    plt.show()
    

def plot_pr(y_test, predict):  
    p, r, thresholds = precision_recall_curve(y_test, predict)
    
    plt.title('PR')  
    plt.fill_between(r, p, alpha=0.2)  
    plt.grid(True, linestyle='-', color='0.75')  
    plt.plot(r, p, linewidth=2, label="roc")
    plt.plot(r[:-1], thresholds, linewidth=1, label="thresholds")
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.axis([0, 1, 0, 1])
    plt.show()
   
   
```
